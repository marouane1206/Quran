```python
import nltk
import pyquran as q
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
quran_df= pd.DataFrame(columns=["Sura Number", "Sura Name AR","Sura Text","Number of Ayat","Sura Words","Sura Word Count","Sura Letter Count"])
quran_df['Sura Number']=range(1,116)
```

```python
def ayat_count(sura): 
    ayat_count= len(q.quran.get_sura(sura))
    return ayat_count
```

```python
ayat_count(22)
```

```python
def sura_words(sura):
    n=0   
    total = 0
    for n in range(len(q.quran.get_sura(sura))):
        total=len(q.quran.get_verse(sura,n).split()) + total  
    return total
```

```python
sura_words(114)
```

```python
def aya_words(sura,aya):
    return len(q.quran.get_verse(sura,aya).split())
```

```python
aya_words(114,1)
```

```python
def sura_letters(sura):
    n=0   
    total = 0
    for n in range(len(q.quran.get_sura(sura))):
        total = len(list(q.quran.get_verse(sura,n))) + total
    return total
```

```python
sura_letters(114)
```

```python
def aya_letters(sura,aya):
    total=0
    n=0 
    for n in range(len(q.quran.get_verse(sura,aya))):        
        total = len(list(q.quran.get_verse(sura,aya)))
    return total
```

```python
aya_letters(114,1)
```

```python
for s in quran_df['Sura Number'].index:
    quran_df['Sura Number'][s] = s
    quran_df['Sura Name AR'][s] = q.quran.get_sura_name(s)
    quran_df["Sura Text"][s] = q.quran.get_sura(s)
    quran_df['Number of Ayat'][s] = ayat_count(s)
    quran_df["Sura Words"][s] = list(q.quran.get_sura(s))
    quran_df['Sura Word Count'][s] = sura_words(s)
    quran_df['Sura Letter Count'][s] =sura_letters(s)  
```

```python
def sura_df(s):
    ayat_df = pd.DataFrame(q.quran.get_sura(s),columns=["Aya Text"])
    ayat_df['Sura Number'] = s
    ayat_df['Sura Name'] = q.quran.get_sura_name(s)
    ayat_df['Aya Number']=ayat_df.index+1
    ayat_df['Number of Ayat']= ayat_count(s)
    ayat_df['Aya Words'] = ayat_df['Aya Text']
    ayat_df['Aya Word Count']=len(q.quran.get_sura(s))
    for n in ayat_df['Aya Text'].index:
        ayat_df['Aya Words'][n] = ayat_df['Aya Text'][n].split()
        ayat_df['Aya Word Count'][n]=aya_words(s,n+1)
    ayat_df['Aya Letter Count']=ayat_df['Aya Text'].apply(len)
    return ayat_df
```

```python
def get_aya(s,n):
    return quran_all_df[(quran_all_df['Sura Number']==s) & (quran_all_df['Aya Number']==n)]
```

```python
def change_place(s,n,place):
    i= get_aya(s,n).index
    quran_all_df.at[i,'Place']= place
    return get_aya(s,n)
```

```python
def bulk_change(s,a,b,place):
    for i in range(a,b+1):
        changed = change_place(s,i,place)
    return changed
```
